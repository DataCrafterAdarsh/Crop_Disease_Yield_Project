```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import matplotlib
matplotlib.rcParams['figure.dpi'] = 80
```

```python
data = pd.read_csv("../data/yield_data_cleaned.csv")
data.head()
```

```python
data.describe()
```

```python
plt.figure(figsize=(6,4))
plt.title("Distribution of Crop Yield")
plt.xlabel("Yield")
plt.ylabel("Frequency")
plt.show()
```

```python
sampled_data = data.sample(3000, random_state=42)

plt.figure(figsize=(6,4))
plt.scatter(sampled_data['annual_rainfall'], sampled_data['yield'], s=10)
plt.xlabel("annual_rainfall")
plt.ylabel("Yield")
plt.title("annual_rainfall vs Yield (Sampled)")
plt.show()
```

```python
top_crops = data['crop'].value_counts().index[:5]
filtered_data = data[data['crop'].isin(top_crops)]

plt.figure(figsize=(8,4))
sns.boxplot(x=filtered_data['crop'], y=filtered_data['yield'])
plt.xticks(rotation=45)
plt.title("Yield for Top 5 Crops")
plt.show()
```

```python
corr = data.select_dtypes(include='number').corr()

plt.figure(figsize=(6,5))
sns.heatmap(corr, cmap="coolwarm")
plt.title("Feature Correlation Heatmap")
plt.show()
```

```python
corr = data.select_dtypes(include='number').corr()

plt.figure(figsize=(8,6))
sns.heatmap(corr, cmap="coolwarm")
plt.savefig("../results/yield_correlation_heatmap.png")
```

```python
Exploratory Data Analysis was performed to understand relationships among key variables
affecting crop yield. 
Visualizations such as histograms, scatter plots, box plots, and correlation heatmaps were used.
Rainfall and cultivated area showed a noticeable influence on crop yield, 
confirming their importance for prediction.
```
